# 05 — Контакт двух пластин (A4)

Круг (q₁=4) над кольцом (q₂=1), касание Δ=0. Case: `cases/ci/two_plates_ring.toml`. Реакция — пара ±r; над дыркой кольца r ≡ 0.

```python
import matplotlib.pyplot as plt
import numpy as np

from plate_solver.dispatch import solve
from plate_solver.problem import Problem

res = solve(Problem.from_toml("../cases/ci/two_plates_ring.toml"))
c = res.contact
fig, axes = plt.subplots(1, 3, figsize=(16, 4.6))
for ax, (fld, ttl) in zip(axes, [(c.w_grid, "w₁ (круг)"),
                                 (c.w2_grid, "w₂ (кольцо)"),
                                 (c.r_grid, "реакция r (пара ±r)")]):
    pcm = ax.pcolormesh(c.Xg, c.Yg, np.ma.masked_invalid(fld), shading="auto")
    ax.set_aspect("equal")
    ax.set_title(ttl)
    fig.colorbar(pcm, ax=ax)
```

Профиль реакции вдоль оси x: контакт только на пересечении планформ.

```python
q1 = c.plate.quad
mask = np.abs(q1.y) < 0.03
order = np.argsort(q1.x[mask])
plt.plot(q1.x[mask][order], c.r_nodes[mask][order], "o-", ms=3)
plt.xlabel("x"); plt.ylabel("r"); plt.grid(alpha=0.3)
plt.title("Профиль r вдоль y≈0: над дыркой кольца r ≡ 0")
```
